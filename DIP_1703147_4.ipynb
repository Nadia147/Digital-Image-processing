**Edge Detcation Of an Image**

```python


img=cv2.imread('/content/drive/MyDrive/Colab Notebooks/DIP Lab Manual 3_files/image003.jpg')
from matplotlib import pyplot as plt

cv2_imshow(img)

#img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
r,c,b=img.shape
print(r)
print(c)
imgb=np.zeros([r+2,c+2,3],dtype=np.int16)
imgb.fill(0)
for i in range(r):
  for j in range(c):
    imgb[i+1][j+1]=img[i][j]





for i in range(1,r):
  for j in range(1,c):
   img[i][j]=imgb[i][j]*(-1)+imgb[i-1][j]*(1)+imgb[i][j-1]*(-1)+imgb[i+1][j]*(1)+imgb[i][j+1]*(-1)+imgb[i-1][j-1]*(1)+imgb[i+1][j+1]*(1)+imgb[i-1][j+1]*(1)+imgb[i+1][j-1]*(1)
        

cv2_imshow(img)
```

**Histogram**

```python

import cv2
  
from matplotlib import pyplot as plt
r,c,v=img.shape

h=np.zeros(256)
for i in range(r):
  for j in range(c):
    h[img[i][j]]+=1
       
x=np.arange(0,256,1)
plt.plot(x,h)

#plt.plot(histr)
plt.show()
```

```python
r,c,v=img.shape
s=43
for i in range(r):
  for j in range(c):
    for k in range(3):
     if img[i][j][k]>s:
      img[i][j][k]=255
     else:
      img[i][j][k]=0
cv2_imshow(img)      
```
